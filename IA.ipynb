<h2>Add Import</h2>

```python
import pandas as pd
import tarfile as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Load datasets</h2>

```python
#Load dataset housing

# path folder
folder = "datasets/"
# path to the archives .tgz
file_tgz = folder+"housing.tgz"

def secure_filter(tarinfo, mode=None):
    """
    Filter to secure the extraction of the archive
    """
    if not tarinfo.name.startswith(("/", "..")):
        return tarinfo
    return None

try:
    # Extraction
    # r:gz = mode d'extraction
    with tf.open(file_tgz, "r:gz") as archive:
        archive.extractall(path=folder,filter=secure_filter)
        print(f"Fichiers extraits")
except OSError as e:
    print(f"Extraction Error : {str(e)}")

data = []
try:
    data = pd.read_csv(folder+"housing.csv")
except FileNotFoundError:
    print(f"CSV file not found")
```

<h2>Imputation missing data</h2>

```python
# Create a DataFrame from the dataset
df = pd.DataFrame(data)

#How many missing values in the dataset
print("DataFrame avant imputation :")
print(f"Il manque {df.isnull().sum().sum()} valeurs dans le dataset")

# Complete the missing values of the DataFrame with the median for numerical columns
median = df['total_bedrooms'].median()
df['total_bedrooms'] = df['total_bedrooms'].fillna(median)

#How many missing values in the dataset
print("DataFrame après imputation :")
print(f"Il manque {df.isnull().sum().sum()} valeurs dans le dataset")
```

<h2>Visualisation data</h2>

<h3>Histogramme</h3>

```python
sns.histplot(df['median_house_value'], bins=50, kde=True)
plt.title("Distribution des prix des maisons")
plt.show()
```

<h3>Diagrammes en barres</h3>

<h3>Cartes de chaleur pour corrélations</h3>

```python
columns = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()
```

<h3>Cartes géographiques</h3>

```python
plt.scatter(df['longitude'], df['latitude'], alpha=0.4, 
            c=df['median_house_value'], cmap='viridis', s=10)
plt.colorbar(label='Prix des maisons')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Visualisation géographique des prix des maisons")
plt.show()
```

<h2>Implementation K-means algorithms</h2>

```python
from sklearn.cluster import KMeans
# Sélection des variables pertinentes
X = df[['longitude', 'latitude', 'median_house_value']]

# Normalisation des données
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Application de K-means avec k=2
kmeans = KMeans(n_clusters=2, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Visualisation des clusters
plt.scatter(df['longitude'], df['latitude'], c=df['cluster'], cmap='viridis', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Clustering des prix des maisons")
plt.show()
```
